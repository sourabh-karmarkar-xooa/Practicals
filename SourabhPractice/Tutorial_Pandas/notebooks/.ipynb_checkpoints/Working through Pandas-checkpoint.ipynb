### Add a new series to df, namely "cat1" where cat1 equals "high" when A and D both are > 0 else "low"

```python
df4 = pd.DataFrame(np.random.randn(15, 4), columns=['A', 'B', 'C', 'D'])

def check_val(row):
    if row[0] > 0 and row[3] > 0:
        return 'high'
    else:
        return 'low'
df4['cat1'] = df4.apply(lambda row:check_val(row),axis=1)

"""
# This also  works

df4['cat1'] = df4.apply(lambda row: 'high' if row[0] > 0 and row[1] > 0 else 'low',axis=1)
"""
print(df4)
```

### Groups and Troups

For all different types of cat1, Find the third largest value of C

```python
# N largest value
N=5
for name,val in df4.groupby('cat1'):
    try:
        print(name, ' ---> ', list(val['C'].nlargest(N))[N-1])
    except:
        print("array doesn't have that many values")
```

### Project Gutenberg (The return of Sherlock Holmes)
- Download the text file of the book
- Create a counter dict of all the words. (use collections.Counter)
- Populate a DataFrame
#### -- What is most occurring word
#### -- What is the least occurring word
#### -- What is the top 10% occurring words (Use a pandas function)
#### -- How many times do the words elementary, holmes, and watson occur ?
#### -- Populate a 4gram. does "elementary_my_dear_watson" occur ? (Punctuations ??)

```python
from collections import Counter
story=Counter(open('sherlockholmes.txt','r').read().split(" "))
story_df=pd.DataFrame({'WORDS':list(story.keys()),'FREQUENCY':list(story.values())},columns=['WORDS','FREQUENCY'])

#print(story_df[story_df['FREQUENCY'] == story_df['FREQUENCY'].max()])
#print((story_df[story_df['FREQUENCY'] == story_df['FREQUENCY'].min()]).sample(5))

# top 10% most occuring words
#print(story_df['FREQUENCY'].quantile(0.9))
sorted_values = story_df.sort_values("FREQUENCY")
sorted_values.iloc[int(21181 * 0.9)]

# plotting on a graph
plot_data=story_df.sample(10)
print(plot_data)
sns.barplot(data=plot_data,x='WORDS',y='FREQUENCY')
```

### Stackoverflow developer survey

Lets replicate the results

Survey results: https://insights.sy times do the words elementary, holmes, and watson occur ?tackoverflow.com/survey/2018/


Get the data from here: https://insights.stackoverflow.com/survey/

```python
survey=pd.read_csv('survey_results_public.csv',delimiter=",")

# percentage of full time students in India
print('Percentage of full-time students in India : ',(survey[(survey.Student == 'Yes, full-time') & (survey.Country == 'India')].shape[0]))

# percentage of full time students country wise
print("Percentage of full time students country wise")

def country_func(con):
    for name,val in con:
        students=val[(val.Student == 'Yes, full-time')]
        percentage=(len(students)/len(val))*100
        yield (name, len(students), percentage)
        # print(name,'------>',len(students),'------> Percentage is ',)
        
pd.DataFrame(country_func(survey.groupby('Country')),columns=['Country','Number','Full-Time Students %'])
```
